## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

## Research Question

**Decision**

Which content pages should be reviewed first for a content refresh?

**Action**

A content reviewer can update, improve, expand, or monitor the highest-ranked pages instead of checking every page manually.

**Cost of a wrong recommendation**

If an important page is ranked too low, the team may miss an opportunity to improve valuable content. If an unimportant page is ranked too high, reviewers waste time reviewing pages with little potential.

**Why ML can help**

Machine learning can combine multiple observable signals such as impressions, CTR, content age, and engagement to prioritize pages more effectively than a simple hand-written rule.
